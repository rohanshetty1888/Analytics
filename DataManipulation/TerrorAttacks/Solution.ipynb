```python
# Q.7) Label all the incidents where the number killed was more than 5 as severe.
#if nkill is greater than 5, then severe otherwise not severe
def severe_not(x):
    if x>5:
        return "Severe"
    else:
        return "Not Severe"
```

```python
#A7=terror['nkill'].map(severe_not)
terror['Terror_Type']=terror['nkill'].map(severe_not)
terror[['nkill','Terror_Type']].sort_values(by='nkill',ascending=False)
```

```python
# Q.8) write a function to label an incident that was both successful and suicidal
def suicide_status(row):
    if (row['suicide']==1 and row['success']==1):
        return 'Successful'
    else:
        return 'Failure'
```

```python
terror['suicide_status']=terror.apply(suicide_status,axis=1)
```

```python
#terror[['suicide','success','suicide_status']]
terror.query("suicide=='1' and success=='1'")[['suicide','success','suicide_status']].head()
```

```python
#terror.query("suicide=='1' and success=='0'")[['suicide','success','suicide_status']].head()
terror.query("suicide=='0' and success=='1'")[['suicide','success','suicide_status']].head()
```

```python
# Q.9) Create a new category representing if the incident occured in Afghanistan, Pakistan or India 
## as one level of the category and all the other countries as another level. 
## Label all other countries as ROW and new column which contains the new category as â€˜Localâ€™ in the data.
def label_country(row):
    if row['country_txt']=="India" or row['country_txt']=="Pakistan" or row['country_txt']=="Afghanistan":
        return "Af-Pak-India"
    else:
        return "ROW"
```

```python
terror['Local']=terror.apply(label_country,axis=1)
```

```python
terror[['country_txt','Local']]
```
